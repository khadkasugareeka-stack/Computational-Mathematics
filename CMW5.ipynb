b) Difference between NumPy array and Python list

Ans:
NumPy arrays hold data of a single type, unlike Python lists which can contain various data types.
Using NumPy necessitates importing the numpy library, whereas lists are built-in. NumPy arrays consume less memory than lists.
Furthermore, NumPy offers superior speed and efficiency.
NumPy facilitates vectorized operations, a capability not present in lists.

```python
#5) Using NumPy for values and Matplotlib for plotting

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = 4*x**2 + 2

plt.plot(x, y,)
plt.title("y = 4x^2 + 2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#6.a) Using Matplotlib 3D projection

import numpy as np
import matplotlib.pyplot as plt

U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0,0,0, U[0],U[1],U[2])
ax.quiver(0,0,0, V[0],V[1],V[2])
ax.quiver(0,0,0, W[0],W[1],W[2])

plt.show()
```

```python
#b) Adding vectors using NumPy and plotting

U = np.array([-0.5, 0.5])
V = np.array([1,1])
W = U + V

import matplotlib.pyplot as plt

plt.quiver(0,0,U[0],U[1])
plt.quiver(0,0,V[0],V[1])
plt.quiver(0,0,W[0],W[1])

plt.xlim(-1,2)
plt.ylim(-1,2)
plt.grid()
plt.show()
```

```python
#c) Magnitude of vector
#formula √(x² + y² + z²)

import numpy as np

def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5,1.5,2.0])
print(magnitude(U))
```

```python
#d) Direction cosines

import numpy as np

def direction_cosines(v):
    mag = np.sqrt(np.sum(v**2))
    return v / mag

O = np.array([1.5,1.5,2.0])
print(direction_cosines(O))
```
