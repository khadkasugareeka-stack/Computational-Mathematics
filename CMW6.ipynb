```python
#Problem 2: Given a matrix and a vector, write a Python function that performs Matrix-Vector product.
#Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-Matrix product.

import numpy as np

def dot_product(vec1, vec2):

    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))

def matrix_vector_product(mat, vec):
    m, n = mat.shape
    p, q = vec.shape

    assert n == p, "The matrix and vector are not compatible for multiplication!"

    result = np.zeros((m, q), dtype=float)

    for i in range(m):
        result[i] = dot_product(mat[i, :], vec)

    return result

mat1 = np.full((3, 3), 3, dtype=int)
vec1 = np.full((3, 1), 2, dtype=int)

print("Matrix = \n", mat1)
print("Vector = \n", vec1)

result1 = matrix_vector_product(mat1, vec1)
print("Result1 (Manual) = \n", result1)

result2 = mat1 @ vec1
print("Result2 (NumPy @) = \n", result2)
```

```python
#Problem 3: Given two compatible matrices, write a Python function that performs Matrix-Matrix product.
#Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-Matrix product.

def matrix_matrix_product(matA, matB):
    m, n = matA.shape
    p, q = matB.shape

    assert n == p, "Matrices are not compatible for multiplication!"

    result = np.zeros((m, q), dtype=float)

    for i in range(m):
        for j in range(q):
            result[i, j] = dot_product(matA[i, :], matB[:, j])

    return result

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

res_manual = matrix_matrix_product(A, B)

res_numpy = A @ B

print("Manual Result:\n", res_manual)
print("NumPy Result:\n", res_numpy)
```

```python
#Problem 4: : Given a matrix, write a Python function that checks if it is a symmetric matrix.

def is_symmetric(X):
    X = np.array(X)
    return np.array_equal(X, X.T)


X = [[9,3,6],
     [2,8,6],
     [3,5,9]]

print(is_symmetric(X))
```
