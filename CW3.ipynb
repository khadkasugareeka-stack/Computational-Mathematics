```python
import matplotlib.pyplot as plt
import math

x = []
y = []

i = -4 * math.pi
while i <= 4 * math.pi:
    x.append(i)
    y.append(math.sin(i))
    i += 0.1

plt.plot(x, y)
plt.title("y = sin(x)")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

r = 4
a, b = 2, 2

theta = 0
while theta <= 2*math.pi:
    x.append(r * math.cos(theta) + a)
    y.append(r * math.sin(theta) + b)
    theta += 0.1

plt.plot(x, y)
plt.scatter(a, b)  # center point

plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

a, b = 3, 5
p, q = 2, 2

theta = 0
while theta <= 2*math.pi:
    x.append(a * math.cos(theta) + p)
    y.append(b * math.sin(theta) + q)
    theta += 0.1

plt.plot(x, y)
plt.scatter(p, q)

plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x_vals = []
y_vals = []

p, q = 0, 0
a = 0.5

y = -5
while y <= 5:
    x = 4*a*(y-p)**2 + q
    x_vals.append(x)
    y_vals.append(y)
    y += 0.1

plt.plot(x_vals, y_vals)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = list(range(20))

y1 = [5*i**3 + 2*i - 1 for i in x]
y2 = [-2*i**3 + i**2 + 100 for i in x]
y3 = [2*math.pi*i + 20 for i in x]

plt.plot(x, y1, label="f1")
plt.plot(x, y2, label="f2")
plt.plot(x, y3, label="f3")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x1, y1 = [], []
x2, y2 = [], []

theta = 0
while theta <= 2*math.pi:
    x1.append(4*math.cos(theta)+2)
    y1.append(4*math.sin(theta)+2)

    x2.append(3*math.cos(theta)+2)
    y2.append(5*math.sin(theta)+2)

    theta += 0.1

plt.plot(x1, y1, label="Circle")
plt.plot(x2, y2, label="Ellipse")

plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```
