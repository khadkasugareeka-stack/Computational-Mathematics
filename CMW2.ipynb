```python
def unique_list(lst):
    unq = []
    for item in lst:
      if item not in unq:
        unq.append(item)
    return unq
sample = [6,8,7,9,0,6,5,7,8,9]
print("Unique list",unique_list(sample))
```

```python
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
n=int(input("Enter a number "))
if prime(n):
    print("{} is a prime number".format(n))
else:
  print("{} is not a prime number".format(n))
```

```python
def twin_prime(limit=1000):
    for i in range( 3, limit - 2):
        if prime(i) and prime(i + 2):
            print(i, "&", i + 2)
twin_prime()
```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n + 1):
        if n % i == 0:
            total += i
    return total

def next_perfect_number(m):
    num = m
    while True:
        if sumOfDivisors(num) == 2 * num:
            return num
        num += 1

print("Next perfect number",next_perfect_number(7))
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for i in marks:
    total += i
mean = total / len(marks)

# Median
n = len(marks)
if n % 2 == 1:
    median = marks[n // 2]
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2

print("Mean:", mean)
print("Median:", median)
```

```python
def decimal_to_binary(num):
    binary = ""
    while num > 0:
        remainder = num % 2
        binary = str(remainder) + binary
        num = num // 2
    return binary

# Example
print(decimal_to_binary(40))
```

```python
def prodDigits(num):
    product = 1
    while num > 0:
        digit = num % 10
        product *= digit
        num = num // 10
    return product

# Example
print(prodDigits(1234))
```

```python
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# a. Food1,2,3,4,5
print("Result (a):", multiply_all(1, 2, 3, 4, 5))

# b. 12,13,14
print("Result (b):", multiply_all(12, 13, 14))
```

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Length of set1
print("a. Length of set1:", len(set1))

# b. Sum of elements in set2
print("b. Sum of set2:", sum(set2))

# c. Union
print("c. Union:", set1 | set2)

# d. Intersection
print("d. Intersection:", set1 & set2)

# e. set1 - set2
print("e. set1 - set2:", set1 - set2)

# f. set2 - set1
print("f. set2 - set1:", set2 - set1)

# g. Check subset
print("g. set3 subset of set1:", set3.issubset(set1))
print("   set3 subset of set2:", set3.issubset(set2))

# h. Complement of set1 (U = {1 to 50})
U = set(range(1, 51))
print("h. Complement of set1:", U - set1)

# i. Check if 15 is present in set1
print("i. Is 15 in set1?", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("j. After adding 60 to set1:", set1)

# k. Remove 60 from set2
set2.discard(60)   # discard avoids error if not present
print("k. After removing 60 from set2:", set2)

# l. Remove all elements from set3
set3.clear()
print("l. set3 after clearing:", set3)

# m. Max and Min in set2
print("m. Max in set2:", max(set2))
print("   Min in set2:", min(set2))
```
